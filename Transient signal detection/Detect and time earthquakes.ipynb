```python
# Find peaks in the charactristic function (detecting earthquakes) + Send back peaks to P arrival times
# Outputs are: 1- marker file of detected earthquakes 2- marker file of P arrival times

import numpy as np
import obspy
from obspy import UTCDateTime, read, Trace, Stream
import math
import pandas
import pandas as pd
from numpy import *
import datetime
import librosa

file = ('2014251-characteristic function.txt')
N =np.loadtxt (file)

time= '2014-09-08'

# Normalize charactristic function according to its maximum and minimum
#N= normalized = (o-min(o))/(max(o)-min(o))

p=input('peak=' )  # Ask for the thereshold of picking peaks
#p= 0.002
# Find peaks in the charactristic functin
peaks = librosa.util.peak_pick(N, 90, 35, 35, 35, float(p), 32)

# Save peaks based on sample numbers
with open ('peak.txt', 'w')as outfile:
    outfile.write("\n".join(map(lambda x: '%i' % x, peaks)))
    
 # Convert sample numbers to time (seconds)
x= loadtxt ('peak.txt')
for i in x:
    with open ('peak.txt', 'w')as outfile:
        t=(x-1)*0.32
        d=pd.to_timedelta (t, unit='s')
        outfile.write("\n".join(map(lambda x: '%s' % x, d)))
        
# Create the final marker file of peaks (detected eqrthquakes)
header= '# Snuffler Markers File Version 0.2'
data= open("peak.txt","r").readlines()

output = ["%s %s  %s %s" % ('2014-09-08', item.split(" ")[2].strip(), 1, '7Z.HOLU..HHZ') for item in data]
f = open('EQ detected' + '2014-09-08p' + str (p)+ '.txt',"w")
f.write(header + "\n")
f.write("\n".join(output))
f.close()

# An algorithm for sending back the peaks to P arrival times
# In this algorithm we select a range preceding each peak and make this range smaller in each step. Finally we find the changing point in the remaining range

result = []
for j in range (len (peaks)):
    peak = peaks [j]

    # Cut the data to a range (start, end) of each peak and some previous seconds (4 seconds here)
    start=peak-19 # 19 samples =5 seconds (19 samples ~18.75 = [5*3.125(sr) + 3.125]. 1* 3.125 is for considering 1 second before the onset in the range)
    end=peak      # When peak is the end of the range it will calculate until the sample of peak-1

    # Set values of y less than 0.0001 (in not normalized data) to 0.0 (These small numbers are not signs of transient signals)

    for z in range (start,end):
        N[N < 0.000001] = 0.0
        #print (y[z])
    
    # If in the (start,end) period there is 4 continuous y samples (1.6 seconds) with zero values we skip that part and start from the first non-zero value after 5 zeros

    for l in range (start,end):
        if N[l]==N[l-1]==N[l-2]==N[l-3]==0.0 and N[l+1]>0:
            start=l
            #print ('start=',l)
        else:
            start=start
            #print ('zero=',start)
        
    # Find local maximums preceding the peak and skip the samples which comes after that

    mylst = list()
    for i in range (start,end):
        if N[i-1]<N[i]>N[i+1] and N[i]>0.000468:
            mylst.append(i)
    if len (mylst) != 0:
        stop1 =min (mylst)
        #print ('min1=', stop1)
    else:
        stop1=end
        #print ('end1=',end)
        
    # Skip the samples which comes after a point with this situation: the slope of the right sideof a point is less than the left side slope

    mylst1 = list()
    for s in range (start,stop1):
        if N[s]-N[s-1]>N[s+1]-N[s]:
            mylst1.append(s)
    if len (mylst1) != 0:
        stop =min (mylst1)
        #print ('min=', stop)
    else:
        stop=stop1
        #print ('end=',end)

    # Calculate the relative of slopes in both sides of each points in the remaining range

    mylst2 = list()
    for k in range (start, stop):
        slop= (N[k+1]-N[k])/(N[k]-N[k-1])
        mylst2.append(slop)
        
        if slop == nan:
            start= k
        else:
            start=start
    
    # Find the maximum slope (which is the minimum angel) and the index of that in the data
    # The condition val<inf is defined here to ignore not defined 'inf' slopes
    # If there is only one value remained and that is 'inf' then the onset is the index of that value
    # If mylst2 = [nan, inf], we should take next sample after the start. For this situation we write: len (mylst2)==2 and mylst2[1]==inf
    
    if len (mylst2)==1:
        idx= 1
    elif len (mylst2)==0:
        idx=0
    elif len (mylst2)==2:
        idx=1
    elif mylst2[:]==inf:
        idx=1
    elif mylst2[:]==nan:
        idx=1
        #print (mylst2)
    
    else:
    
        val, idx = max((val, idx) for (idx, val) in enumerate(mylst2) if val<inf)  
        #print (val)
    onset= idx+start
    result.append(onset)

    
# Save peaks based on sample numbers
with open ('peak.txt', 'w')as outfile:
    outfile.write("\n".join(map(lambda x: '%i' % x, result)))
    
 # Convert sample numbers to time (seconds)
b= loadtxt ('peak.txt')
for i in x:
    with open ('peak.txt', 'w')as outfile:
        t2=(b-1)*0.32
        d=pd.to_timedelta (t2, unit='s')
        outfile.write("\n".join(map(lambda x: '%s' % x, d)))
        
# Create the final marker file of P arrival times
header= '# Snuffler Markers File Version 0.2'
data= open("peak.txt","r").readlines()

output = ["%s %s  %s %s" % ('2014-09-08', item.split(" ")[2].strip(), 0, '7Z.HOLU..HHZ') for item in data]
f = open('Parrival' + '2014-09-08p' + str (p) + '.txt' ,"w")
f.write(header + "\n")
f.write("\n".join(output))
f.close()
```
